# Calculating Happiness Score for Student Groups
## Input Parameters used
1. Preferred student 1
1. Preferred student 2
1. Rejected student
1. Max. number of students per group

```python
#https://raw.githubusercontent.com/bettylucy/Practica1/main/StudentPreferences.csv
import pandas as pd
import numpy as np
import math
import itertools
```

```python
#Reading file / Data Validation

SF=pd.read_csv("https://raw.githubusercontent.com/bettylucy/Practica1/main/StudentPreferences.csv")
SF.info()
SF.head()
```

```python
#Asking for maximum number of students per group
maxs=input("Number of students per group: ")
```

```python
#Calculating and generating seating table with neighbors
tablesnum=len(SF)//int(maxs)
tables=[]
for i in range(1,tablesnum+1):
    tables.append(i)
```

```python
#Creating dataframe for hard search
results = pd.DataFrame() 
totaliterations=math.comb(len(SF),int(maxs))
iterations=[]
for i in range (1,totaliterations+1):
    iterations.append(i)
results["iteration"]=iterations

names=SF.loc[:,"Student Name"]
combinations=itertools.combinations(names, 3)
comb = [' '.join(i) for i in combinations]

for i in range(int(maxs)):
    names=[]
    for j in range (len(comb)):
        temp=comb[j].split()
        names.append(temp[i])
    column='Student '+str(i)
    results[column]=names

results
```

```python
#Assigning happiness level to each iteration
happiness=[]

for i in range (len(results)):
    happy=0
    for k in range (len(SF)):
        if results.iat[i,1]==SF.iat[k,0]:
            if SF.iat[k,1]==results.iat[i,2] or SF.iat[k,1]==results.iat[i,3]:
                    happy=happy+10
            if SF.iat[k,2]==results.iat[i,2] or SF.iat[k,1]==results.iat[i,3]:
                    happy=happy+7
            if SF.iat[k,3]==results.iat[i,2] or SF.iat[k,1]==results.iat[i,3]:
                    happy=happy-3
            else:
                    happy=happy+1
        if results.iat[i,2]==SF.iat[k,0]:
            if SF.iat[k,1]==results.iat[i,1] or SF.iat[k,1]==results.iat[i,3]:
                    happy=happy+10
            if SF.iat[k,2]==results.iat[i,1] or SF.iat[k,1]==results.iat[i,3]:
                    happy=happy+7
            if SF.iat[k,3]==results.iat[i,1] or SF.iat[k,1]==results.iat[i,3]:
                    happy=happy-3
            else:
                    happy=happy+1
        if results.iat[i,3]==SF.iat[k,0]:
            if SF.iat[k,1]==results.iat[i,2] or SF.iat[k,1]==results.iat[i,1]:
                    happy=happy+10
            if SF.iat[k,2]==results.iat[i,2] or SF.iat[k,1]==results.iat[i,1]:
                    happy=happy+7
            if SF.iat[k,3]==results.iat[i,2] or SF.iat[k,1]==results.iat[i,1]:
                    happy=happy-3
            else:
                    happy=happy+1
    happiness.append(happy)
results["Happiness"]=happiness
results
```

```python
#Obtaining 4 happiest groups

groups=[]
grouphap1=happiness

a=max(grouphap1)
max1=grouphap1.index(a)
groups.append(results.loc[max1,:])
del grouphap1[max1]

grouphap2=grouphap1
```
